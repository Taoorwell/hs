```python
log = 'log3.txt'
pf = open(log, 'w')
for i in [2, 3]:
    for msize in m:
        train_samples, train_labels = get_train_sample(data_path=MAIN_FOLDER+DATA_PATH[i],
                                                       train_data_path=MAIN_FOLDER+TRAIN_PATH[i],
                                                       c=c[i], d=4, lists=listss[i], m=msize)
    
        input_shape = (msize, msize, band[i])
        x_train, x_valid, y_train, y_valid = train_test_split(train_samples, train_labels,
                                                              test_size=0.1, shuffle=True,
                                                              stratify=train_labels)
        y_train, y_valid = one_hot_encode(c=c[i], labels=y_train), one_hot_encode(c=c[i], labels=y_valid)
            
        model = cnn_2d_pca(input_shape=input_shape, c=c[i], lr=0.001, rate1=0.2, rate2=0.2, l=0.0001)
        start = time.clock()
        print("Data: {}, Size of M: {}\n".format((name1[i]), (msize)))
        pf.write("Data: {}, Size of M: {}\n".format((name1[i]), (msize)))
        network = model.fit(x_train, y_train, batch_size=30, epochs=100, verbose=1,
                            validation_data=(x_valid, y_valid))
        end = time.clock()
        time0 = end - start
        print("Training Finish!! Time Consume: {}\n".format(time0))
        pf.write("Training Finish!! Time Consume: {}\n".format(time0))
        save_model(model, "new_model_2/"+str(msize)+"-"+name1[i])
        print("Save Model Sucessfully!\n")
        pf.write("Save Model Sucessfully!\n")
```
